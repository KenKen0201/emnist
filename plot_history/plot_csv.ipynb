```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
```

```python
df02 = pandas.read_csv("training_history_emnist_dropout_0.2.csv")
df03 = pandas.read_csv("training_history_emnist_dropout_0.3.csv")
df04 = pandas.read_csv("training_history_emnist_dropout_0.4.csv")
```

```python
df02 = df02.set_index("Modell")
df03 = df03.set_index("Modell")
df04 = df04.set_index("Modell")
```

```python
# set x label range, starting with 32
r = np.arange(16,288 + 16,16)
r
```

```python
axes = plt.gca()

# plot ensemble accuracy over training history
en_acc = df02.loc["Ensemble",:]
ensemble_accuracy = en_acc.values.tolist()
plt.plot(r, ensemble_accuracy, "bs-")

# plot model accuracies over training history
for i in range(10):
    acc = df02.loc["Modell {}".format(i),:]
    accuracy = acc.values.tolist()
    plt.plot(r, accuracy)

plt.title("0.2 dropout model accuracy over training history")
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(["Ensemble"], loc='upper right')
#plt.xlim(32,224)
plt.ylim(0.16, 0.5)
plt.show()
```
