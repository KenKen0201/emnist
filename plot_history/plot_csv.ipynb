```python
import pandas
import numpy as np
```

```python
df = pandas.read_csv("training_history_emnist_dropout_0.2.csv")
```

```python
df2 = df.set_index("Modell")
df2
```

```python
# set x label range, starting with 32
r = np.arange(32,288 + 16,16)
r
```

```python
import matplotlib.pyplot as plt

axes = plt.gca()

# plot ensemble accuracy over training history
en_acc = df2.loc["Ensemble",:]
ensemble_accuracy = en_acc.values.tolist()
plt.plot(r, ensemble_accuracy, "bs-")

# plot model accuracies over training history
for i in range(10):
    acc = df2.loc["Modell {}".format(i),:]
    accuracy = acc.values.tolist()
    plt.plot(r, accuracy)

plt.title("model accuracy over training history")
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(["Ensemble"], loc='upper left')
#plt.xlim(32,224)
plt.show()
```
